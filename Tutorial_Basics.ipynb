## Conditionals (IF you want to, or ELSE you don't)

Conditionals are used to make certain parts of run if a certain condition is met. Trying change the value of A and seeing what happens

````python
A = "Apple"
if A == "Apple":
    print("It's a Apple")
else:
    print("It's an Orange")
````

## Loop de Loops

Loops are used to repeat a code while a condition has/hasn't been met. Below are an examples of a "while" and "for" Loop

````python
i = 10
while i > 0:
    i = i - 1 
    print("What is it?")
    
print("It's a red and blue striped golfing umbrella")
````

````python
for i in range(10):
    print(i)
````

Sometimes, we want to define our own functions. In Python, instead of doing:
```idl
function hypotenuse, shortsideone, shortsidetwo
    longside = ((shortsideone)^2+(shortsidetwo)^2)^(0.5)
    return, longside
end
```
We use:
```python
def myfunction():
    return something1, something2, something3
```
And we are not restricted to only returning one variable!!

````python
#This is the function, note the indented lines are what will run when the function is called
#shortsideone and shortsidetwo are the inputs of the function

def hypotenuse(shortsideone,shortsidetwo):
    longside = ((shortsideone)**2+(shortsidetwo)**2)**(0.5)
    return longside

a = 5
b = 6

# Now we call the function to make it run, using the values we defined 
hypotenuse(a,b)
````
